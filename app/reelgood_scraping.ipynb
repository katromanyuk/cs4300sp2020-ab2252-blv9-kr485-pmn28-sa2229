```python
sry = 'Sorry, nothing to see here! Try something else.'
def get_titles(service):
    shows = []
    html = '' 
    offset = 0
    while sry not in html:
        url = 'https://www.w3.org/services/html2txt?url=https%3A%2F%2Freelgood.com%2Fmovies'\
        +'%2Fsource%2F'+service+'%3Ffilter-sort%3D4%26offset%3D'+str(offset)+'&nonums=on'
        html = requests.get(url).text.replace('\n', ' ')
        lst = re.findall('     \[(.*?)https://reelgood.com/movie', html)
        shows += [s.strip() for s in lst]
        offset+=50
    return shows
```

```python
movielist = {}

def combined():
    for x in netflix:
        movielist[x] = [1,0,0,0,0,0]
    for x in hulu:
        if x in movielist:
            movielist[x][1]+=1
        else:
            movielist[x] = [0,1,0,0,0,0]
    for x in amazon:
        if x in movielist:
            movielist[x][2]+=1
        else:
            movielist[x] = [0,0,1,0,0,0]
    for x in disney:
        if x in movielist:
            movielist[x][3]+=1
        else:
            movielist[x] = [0,0,0,1,0,0]
    for x in apple:
        if x in movielist:
            movielist[x][4]+=1
        else:
            movielist[x] = [0,0,0,0,1,0]
    for x in hbo:
        if x in movielist:
            movielist[x][5]+=1
        else:
            movielist[x] = [0,0,0,0,0,1]

combined()
```

```python
moviedic = {}

def mergelist(lst1, lst2):
    lst = [0,0,0,0,0,0]
    for x in range(6):
        if lst1[x] == 1 or lst2[x]==1:
            lst[x]+=1
    return lst


def breakdown(st):
    st = st.split(' ')
    rotten = st[-1]
    imdb = st[-2]
    age = st[-3]
    year = st[-4]
    name = " ".join(st[:-4])
    return ([name, year, age, imdb, rotten])

for x in movielist:
    val = breakdown(x)
    if val[0] in moviedic:
        moviedic[val[0]]['services'] = mergelist(moviedic[val[0]]['services'], movielist[x])
    else:
         moviedic[val[0]] = {'year': val[1], 'age': val[2], 'imdb':val[3], 'rotten': val[4], 'services': movielist[x]}
   
```

```python
omdb_TOKEN = '4292bf53'
def find_movie(movie):
    plot = ""
    query = "http://www.omdbapi.com/?apikey=" + omdb_TOKEN + "&t=" + movie
    params = {"r": "json", "plot": "full"}
    result = requests.get(query, params)
    if response(result.text):
        plot = cleanjson(result.text)
    else:
        return "ERROR"
    return (plot)


def cleanjson(result):
    plot = result[result.find("Plot")+7:result.find("Language")-3]
    return plot

def response(result):
    text = result[result.find('"Response":')+12:]
    return text.find('True') > -1
```
