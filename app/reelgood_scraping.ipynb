```python
import re
import requests
```

```python
sry = 'Sorry, nothing to see here! Try something else.'
def get_titles(service):
    shows = []
    html = '' 
    offset = 0
    while sry not in html:
        url = 'https://www.w3.org/services/html2txt?url=https%3A%2F%2Freelgood.com%2Fmovies'\
        +'%2Fsource%2F'+service+'%3Ffilter-sort%3D4%26offset%3D'+str(offset)+'&nonums=on'
        html = requests.get(url).text.replace('\n', ' ')
        lst = re.findall('     \[(.*?)https://reelgood.com/movie', html)
        shows += [s.strip() for s in lst]
        offset+=50
    return shows
```

```python
netflix = get_titles('netflix')
```

```python
hulu = get_titles('hulu')
```

```python
amazon = get_titles('amazon')
```

```python
print(len(netflix))
print(len(hulu))
print(len(amazon))
```
