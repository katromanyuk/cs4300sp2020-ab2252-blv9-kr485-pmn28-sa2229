```python
import re
import requests
import pandas as pd
```

```python
sry = 'Sorry, nothing to see here! Try something else.'
def get_titles(service):
    shows = []
    html = '' 
    offset = 0
    while sry not in html:
        url = 'https://www.w3.org/services/html2txt?url=https%3A%2F%2Freelgood.com%2Fmovies'\
        +'%2Fsource%2F'+service+'%3Ffilter-sort%3D4%26offset%3D'+str(offset)+'&nonums=on'
        html = requests.get(url).text.replace('\n', ' ')
        lst = re.findall('     \[(.*?)https://reelgood.com/movie', html)
        shows += [s.strip() for s in lst]
        offset+=50
    return shows
```

```python
netflix = get_titles('netflix')
```

```python
hulu = get_titles('hulu')
```

```python
amazon = get_titles('amazon')
```

```python
movie_data = pd.read_csv('merged_data2.csv')
```

```python
movie_services = {}
for title in movie_data['Title']:
    movie_services[title] = []
```

```python
updated_n = [] 
for line in netflix:
    a = line.find('200')
    b = line.find('201')
    c = line.find('199')
    d = line.find('198')
    if a != -1:
        updated_n.append(line[:a-1])
    if b != -1:
        updated_n.append(line[:b-1])
    if c != -1:
        updated_n.append(line[:c-1])
    if d != -1:
        updated_n.append(line[:d-1])
```

```python
updated_h = [] 
for line in hulu:
    a = line.find('200')
    b = line.find('201')
    c = line.find('199')
    d = line.find('198')
    if a != -1:
        updated_h.append(line[:a-1])
    if b != -1:
        updated_h.append(line[:b-1])
    if c != -1:
        updated_h.append(line[:c-1])
    if d != -1:
        updated_h.append(line[:d-1])
```

```python
updated_a = [] 
for line in amazon:
    a = line.find('200')
    b = line.find('201')
    c = line.find('199')
    d = line.find('198')
    if a != -1:
        updated_a.append(line[:a-1])
    if b != -1:
        updated_n.append(line[:b-1])
    if c != -1:
        updated_a.append(line[:c-1])
    if d != -1:
        updated_a.append(line[:d-1])
```

```python
def insert_netflix(netflix):
    for title in movie_services.keys():
        for line in netflix:
            if title == line:
                (movie_services[title]) = ['Netflix']
    return movie_services
```

```python
insert_netflix(updated_n)
```

```python
def insert_hulu(hulu):
    for title in movie_services.keys():
        for line in hulu:
            if title == line:
                (movie_services[title]).append('Hulu')
    return movie_services
```

```python
insert_hulu(updated_h)
```

```python
def insert_amazon(amazon):
    for title in movie_services.keys():
        for line in amazon:
            if title == line:
                (movie_services[title]).append('Amazon')
    return movie_services
```

```python
insert_amazon(updated_a)
```

```python
new_col = pd.Series([])
for m, movie in enumerate(movie_data['Title']):
    s = movie_services.get(movie)
    if s == [] or s == None:
        new_col[m] = "None"
    else:
        empty = ""
        for elt in s:
            empty = empty + elt + ", "
        new_col[m] = empty[:-2]
movie_data.insert(12, "Streaming Services", new_col)
```

```python
movielist = {}

def combined():

    for x in netflix:
        movielist[x] = [1,0,0]

    for x in hulu:
        if x in movielist:
            movielist[x][1]+=1
        else:
            movielist[x] = [0,1,0]
    
    for x in amazon:
        if x in movielist:
            movielist[x][2]+=1
        else:
            movielist[x] = [0,0,1]

    
combined()
```

```python
for x in movielist:
    print(str(x) + " : " + str(movielist[x]))
```

```python
print(len(movielist))
```

```python
def summovie(lst):
    c = 0
    for x in lst:
        c+=x
    return c


for x in movielist:
    if summovie(movielist[x]) >1:
        print(str(x) + " : " + str(movielist[x]))
```

```python
def breakdown(st):
    lst = st.split(" ")
    rotten = lst[-1]
    imdb = lst[-2]
    age = lst[-3]
    year = lst[-4]
    name = " ".join(lst[:-4])
    
    return [name, year, age, imdb, rotten]
    
```

```python
moviedict = {}
for x in movielist:
    broken = breakdown(x)
    temp = {'year': broken[1], 'age': broken[2], 'imdb': broken[3], 'rotten': broken[4], 'services': movielist[x]}
    moviedict[broken[0]] = temp
```

```python
for x in moviedict:
    print(str(x) + " : " + str(moviedict[x]))
```

omdb_TOKEN = '4292bf53'
def find_movie(movie):
    plot = ""
    query = "http://www.omdbapi.com/?apikey=" + omdb_TOKEN + "&t=" + movie
    params = {"r": "json", "plot": "full"}
    result = requests.get(query, params)
    if response(result.text):
        plot = cleanjson(result.text)
    else:
        return "ERROR"
    return (plot)

def cleanjson(result):
    plot = result[result.find("Plot")+7:result.find("Language")-3]


def response(result):
    text = result[result.find('"Response":')+12:]
    return text.find('True') > -1

```python
for x in moviedict:
    plot = find_movie(x)
    moviedict[x]['plot'] = plot
```
