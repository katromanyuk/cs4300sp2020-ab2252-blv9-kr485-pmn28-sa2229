```python
import re
import requests
```

```python
sry = 'Sorry, nothing to see here! Try something else.'
def get_titles(service):
    shows = []
    html = '' 
    offset = 0
    while sry not in html:
        url = 'https://www.w3.org/services/html2txt?url=https%3A%2F%2Freelgood.com%2Ftv'\
        +'%2Fsource%2F'+service+'%3Ffilter-sort%3D4%26offset%3D'+str(offset)+'&nonums=on'
        html = requests.get(url).text.replace('\n', ' ')
        lst = re.findall('     \[(.*?)https://reelgood.com/show/', html)
        shows += [s.strip() for s in lst]
        offset += 50
    return shows
```

```python
netflix = get_titles('netflix')
```

```python
hulu = get_titles('hulu')
```

```python
amazon = get_titles('amazon')
```

```python
print(len(netflix))
print(len(hulu))
print(len(netflix))
```

```python
import re
import string
import lyricsgenius
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
TOKEN = 'sD0C3epnJdfOQQK4eIC45dHl-Qv7DipToGpuj1n4WeuG5_LDP1HKn31w5Cn1lOux'
genius = lyricsgenius.Genius(TOKEN)
genius.verbose = False
genius.remove_section_headers = True
```

```python
#artist = genius.search_artist("Arctic Monkeys")
#song = genius.search_song("Don't Sit Down", artist.name)
```

```python
song = genius.search_song('Moved Your Chair', 'Arctic Monkeys')
print(song.lyrics)
```

```python
song = genius.search_song('happier', 'bastille')
print(song.lyrics)
```

```python
stop_words = set(stopwords.words('english'))
lyrics = song.lyrics.lower()
lyrics = re.sub(r'\d+', '', lyrics)
lyrics = re.sub(r'-|\'', ' ', lyrics)
lyrics = ''.join(x for x in lyrics if x not in string.punctuation)
tokens = word_tokenize(lyrics)
processed_lyrics = ' '.join(i for i in tokens if not i in stop_words)
print(processed_lyrics)
```

```python
analyzer = SentimentIntensityAnalyzer()
sentiment = analyzer.polarity_scores(processed_lyrics)
print(sentiment)
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('imdb250.csv')
```

```python
data = data[['Title','Genre','Plot']]
```

```python
data
```

```python
genres = {}
for s in data['Genre']:
    s = s.replace(',','')
    for g in s.split():
        if g in genres:
            genres[g] = genres[g] + 1
        else:
            genres[g] = 1
print(genres)
```

```python
for k,v in genres.items():
    print(k+' : '+str(v))
```
